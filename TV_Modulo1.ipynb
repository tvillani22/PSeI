### Ejercicio B
>***Usar un ciclo for para hacer la suma de Riemann por izquierda de la función $y=sin(x)$
 entre 0 y π.***

```python
x, step = np.linspace(0, np.pi, retstep=True)
y = np.sin(x)
sum_R = 0

# Usando el for..
for i in x[:-1]:
    sum_R += np.sin(i)*step
print('Sumatoria Riemann por izquierda: ', sum_R.round(4))

# Grafico..
fig, ax = plt.subplots(1, 1, figsize=(9,6))
ax.plot(x, y, 'tab:red')
ax.bar(x[:-1], y[:-1], align='edge', width=step, color='tab:blue', linewidth=0.3, edgecolor='k');
```

***Haciéndola por izquierda y por derecha, con $x \in [0,\pi/2]$ para que se vea la diferencia...***

```python
# Haciendo por izquierda y derecha (hasta pi/2, para que se vea la diferencia)... 
x, step = np.linspace(0, np.pi/2, 20, retstep=True)
y = np.sin(x)

# Directamente..
R_i = (y[:-1]*step).sum().round(4)
R_d = (y[1:]*step).sum().round(4)

# Grafico...
fig, ax = plt.subplots(1, 2, figsize=(15,5))
titles = ['izquierda ' + str(R_i), 'derecha ' + str(R_d)]
for axx, y_, tit in zip(ax, [y[:-1], y[1:]], titles):
    axx.plot(x,y, 'tab:red')
    axx.bar(x[:-1], y_, align='edge', width=step, color='tab:blue', linewidth=0.3, edgecolor='k')
    axx.set_title('Sumatoria Riemann por ' + tit)
```

### Ejercicio C
>***A partir de una matriz nula de 3x4, generar una matriz de 3x4 donde cada elemento de la misma sea igual a la suma de la fila  y la correspondiente columna que ocupa.***

```python
M = np.zeros((3,4))
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i,j] = i + j
M
# Obs: numpy (como python) indexa desde 0.
```
